```python
import numpy as np
import math
from matplotlib import pyplot as plt 
```

```python
# Load Data
# Check data directory
mI_xt = np.load('matrices/Ixt.npy') 
mI_ty = np.load('matrices/Ity.npy')
reference = np.load('matrices/x_entropy.npy')
```

```python
# Configure Global parameters of the plot
fS = 14
params = {'legend.fontsize': fS,
          'figure.figsize': (7, 7),
         'axes.labelsize': fS,
         'axes.titlesize': fS,
         'xtick.labelsize': fS,
         'ytick.labelsize': fS}
plt.rcParams.update(params)

# Create Figure
f, ax = plt.subplots()
# Axis limits using the reference X entropy value
ax.set( xlim = ( 0, reference ), ylim = ( 0, reference ) )
# Set Axis Labels
ax.set_xlabel("MI(X,T)")
ax.set_ylabel("MI(T,Y)")
# Draw a dashed diagonal line
ax.plot( [0, 1], [0, 1], transform = ax.transAxes, linestyle = 'dashed', color = 'k' ) 

# Plot values 
# Variable for gradient color

c1 = np.linspace(1, 350, 350)
alphaV = [0.4, 0.6, 0.8, 1]
# Variable for marker differentiation
markerSymbols = 'ovsXp*d^<>P8hH'
colors = ["red","green","blue","orange", "cyan", "pink","purple", "brown","gray","olive","black","midnigthblue","palegreen","salmon"]
#maps = ["Reds","Greens","Blues","Oranges"]
# Loop for presenting all the Encoders
numE = math.floor(len(mI_xt)/2)+1
for i in range(numE):
    current_Ixt = mI_xt[i]
    current_Ity = mI_ty[i]
    ax.scatter(current_Ixt, current_Ity, label="Encoder {}".format(i+1), s=c1, marker=markerSymbols[i], c=colors[i],  alpha=None, edgecolors ='k' ) # cmap=maps[i],

# Show the plot
#f.show()
ax.legend()
# Show the plot
f.show()
```
